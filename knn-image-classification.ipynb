```python
import numpy as np 
import cv2
import glob
import os 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from joblib import dump
```

```python
data = []
label = []
# preprocessing 
data_address = glob.glob(r"/kaggle/input/fire-dataset/fire_dataset/*/*.png")
for i, address in enumerate(data_address):
    img = cv2.imread(address)
    if img is None:
        print(f"Error reading image: {address}")
        continue  # 
    img = cv2.resize(img,(64,64))
    img = img / 255.0
    img = img.flatten()

    data.append(img)
    label_split = address.split("/")[-1].split(".")[0]
    label.append(label_split)

    if i % 100 == 0:
        print(i)
    
```

```python
data = np.array(data)

x_train, x_test, y_train, y_test = train_test_split(data,label,test_size = 0.2)
```

```python
clf = KNeighborsClassifier()
clf.fit(x_train , y_train)

dump(clf,"fire_classifier.z")
```

```python
acc = clf.score(x_test, y_test)
print(acc*100)
```
